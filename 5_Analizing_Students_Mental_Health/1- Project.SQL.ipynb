# Analyzing Students' Mental Health

```python
import pandas as pd 
import sqlite3 as sql
```

```python
students = pd.read_csv('students.csv')
students.head()
```

```python
#Connecting to the Database
db_con = sql.connect('students.sql.db') #name of the database
execute_sql = db_con.cursor()

#to_sql method
students.to_sql('students_table',
              con = db_con,
              if_exists ='replace',index=False) #df name to sql table

#Dataframe to SQL file
students_sql = pd.read_sql('''SELECT * 
FROM students_table''',con = db_con)
students_sql
```

```python
# Find the number of international students and their average scores by length of stay, in descending order of length of stay
students_filtered = pd.read_sql('''
SELECT stay,
COUNT(*) AS count_int,
ROUND(AVG(todep), 2) AS average_phq,
ROUND(AVG(tosc), 2) AS average_scs,
ROUND(AVG(toas), 2) AS average_as
FROM students_table
WHERE inter_dom = 'Inter'
GROUP BY stay
ORDER BY stay DESC;''',con=db_con)
students_filtered
```
