```python
import pandas as pd
import sqlite3 as sql
from pandasql import sqldf
import mysql.connector
```

```python
cities = pd.read_csv('cities.csv')
```

```python
countries = pd.read_csv('countries.csv')
```

```python
#Connecting to the Database
db_con = sql.connect('cities.sql.db') #name of the database
execute_sql = db_con.cursor()
```

```python
#Connecting to the Database
db_con_1 = sql.connect('countries.sql.db') #name of the database
execute_sql = db_con_1.cursor()
```

```python
#to_sql method
cities.to_sql('cities_table',
              con = db_con,
              if_exists ='replace',index=False) #df name to sql table
```

```python
#to_sql method
countries.to_sql('countries_table',
              con = db_con_1,
              if_exists ='replace',index=False) #df name to sql table
```

```python
inner_join = pd.read_sql('''
SELECT *
FROM cities_table
INNER JOIN countries_table
ON cities_table.country_code = countries_table.code;
''', con=db_con)
```
